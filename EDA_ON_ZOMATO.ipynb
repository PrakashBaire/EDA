# EDA For Zomato Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('zomato.csv', encoding='latin-1')
df.head()
```

# Finding Missing Values

```python
df.isnull().sum()
```

```python
# Finding EDA using python code
 [features for features in df.columns if df[features].isnull().sum()>1]
```

```python
sns.heatmap(df.isnull(),yticklabels=False, cbar=False,cmap='viridis')
```

```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country, on ='Country Code', how = 'left')
```

```python
final_df.head(2)
```

```python
# To check Data Types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names= final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
#plot pie chart - Top 3 countries that uses zomato
plt.pie(country_val[:3], labels= country_names[:3], autopct= '%1.2f%%')
```

```python
#Observations: Zomato max records or trasactions are from India, after that USA and UK
```

```python
final_df.columns
```

```python
#Used Groupby
rating = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating count'})
```

```python
rating
```

```python
## Observation
1) When rating b/w 4.5 to 4.9----> Excellent
2)When rating b/w 3.4 to 4.0----> very good
3)When rating b/w 3.5 to 9.9----> good
4)When rating b/w 3.0 to 3.4----> average
5)When rating b/w 2.5 to 2.9----> average
6)When rating b/w 2.0 to 2.4----> poor
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']= (12,6)
sns.barplot(x="Aggregate rating", y= "Rating count", data = rating)
```

```python
sns.barplot(x="Aggregate rating", y= "Rating count",hue = 'Rating color' ,data = rating, palette = ['white','Red','Orange', 'Yellow', 'Green', 'DarkGreen'])
```

observations:
1) Not rated count is very high
2) Max number of ratings are b/w 2.5 to 3.4


```python
## Countplot
sns.countplot(x = 'Rating color', data = rating, palette = ['white','Red','Orange', 'Yellow', 'Green', 'DarkGreen'])
```
